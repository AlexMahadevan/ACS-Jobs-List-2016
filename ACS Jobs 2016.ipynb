```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/Users/alexmahadevan/Desktop/Projects/ACS-Jobs/acsjobs.csv")
```

```python
df.head()
```

```python
df['fiveyearchangeJOBS'] = (df.jobs2016 - df.jobs2011) / df.jobs2011
```

```python
df['oneyearchangeJOBS'] = (df.jobs2016 - df.jobs2015) / df.jobs2015
```

```python
df['fiveyearchangeEARNINGS'] = (df.earnings2016 - df.earnings2011) / df.earnings2011
```

```python
df['oneyearchangeEARNINGS'] = (df.earnings2016 - df.earnings2015) / df.earnings2015
```

```python
df.info()
```

```python
df.sort_values("oneyearchangeJOBS" , ascending=False)
```

```python
df['ZfiveyearchangeJOBS'] = (df.fiveyearchangeJOBS - df.fiveyearchangeJOBS.mean()) / df.fiveyearchangeJOBS.std()
```

```python
df['ZoneyearchangeJOBS'] = (df.oneyearchangeJOBS - df.oneyearchangeJOBS.mean()) / df.oneyearchangeJOBS.std()
```

```python
df['ZfiveyearchangeEARNINGS'] = (df.fiveyearchangeEARNINGS - df.fiveyearchangeEARNINGS.mean()) / df.fiveyearchangeEARNINGS.std()
```

```python
df['ZoneyearchangeEARNINGS'] = (df.oneyearchangeEARNINGS - df.oneyearchangeEARNINGS.mean()) / df.oneyearchangeEARNINGS.std()
```

```python
df['ztotaljobs'] = (df.jobs2016 - df.jobs2016.mean()) / df.jobs2016.std()
```

```python
df.info()
```

```python
df['Zscore'] = (df.ztotaljobs + df.ZfiveyearchangeJOBS + df.ZoneyearchangeJOBS + df.ZfiveyearchangeEARNINGS + df.ZoneyearchangeEARNINGS) / 5  
```

```python
df.jobs2016.median()
```

```python
df_sorted = df.sort_values('Zscore' , ascending=False)
```

```python
df_final_large = df_sorted[df_sorted.jobs2016 > 60121.0]
```
